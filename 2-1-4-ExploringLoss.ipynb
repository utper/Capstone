```python
import numpy as np
import matplotlib.pyplot as plt
import math
# Edit these parameters to try different loss measurements. Rerun this cell when
# done. Your Y will be calculated as Y=wX+b, so if w=3, and b=-1, then Y=3x-1.

w = 3
b = -1

x = [-1, 0, 1, 2, 3, 4]
y = [-3, -1, 1, 3, 5, 7]
myY = []
Diff = []

for thisX in x:
  thisY = (w*thisX)+b
  myY.append(thisY)

for a, b in zip(myY, y):
    Diff.append(a - b)

print(f"Real Y is {str(y)}")
print(f"My Y is   {str(myY)}")
print(f"Diff is   {str(Diff)}")

# let's calculate the loss
total_square_error = 0
for i in range(0, len(y)):
  # square_error = (y[i] - myY[i]) ** 2
  square_error = Diff[i]**2
  total_square_error += square_error

print(f"My loss is: {str(math.sqrt(total_square_error))}")

plt.figure(figsize=(8,5))

# Datos originales
plt.scatter(x, y, color='blue', label="Datos reales (y)")
plt.plot(x, myY, color='red', linestyle='dashed', label="Suposición (Y' = 3X - 1)")

# Líneas de diferencia
for i in range(len(x)):
    # crea una línea entre dos puntos:
    # Inicio: (x[i], y[i]), es decir, el punto real.
    # Fin: (x[i], myY[i]), es decir, el punto de la suposición.
    plt.plot([x[i], x[i]], [y[i], myY[i]], color="green", linestyle="dotted")

plt.axhline(0, color='black', linewidth=0.2)
plt.axvline(0, color='black', linewidth=0.2)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Comparación entre valores reales y suposición")
plt.legend()
plt.grid()
plt.show()
```
