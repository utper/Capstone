âœ… La funciÃ³n actualizar(frame) cambia los valores de Y, desplazando la onda seno en tiempo real.

ðŸš€ Este cÃ³digo funciona en Jupyter Notebook, VS Code y otras plataformas, pero no en Colab, ya que Colab no soporta animaciones en tiempo real de Matplotlib. Si estÃ¡s en Colab, podrÃ­amos usar HTML para mostrar la animaciÃ³n.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Crear la figura y el eje
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)  # Valores del eje X
line, = ax.plot(x, np.sin(x))  # LÃ­nea inicial

# FunciÃ³n de actualizaciÃ³n para la animaciÃ³n
def actualizar(frame):
    line.set_ydata(np.sin(x + frame * 0.1))  # Desplaza la onda seno en el tiempo
    return line,

# Configurar la animaciÃ³n con FuncAnimation
ani = animation.FuncAnimation(fig, actualizar, frames=100, interval=50)

plt.title("AnimaciÃ³n de una onda seno")
plt.xlabel("X")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
