üîé ¬øQu√© hace este c√≥digo?

‚úÖ Crea un gr√°fico inicial con una onda seno est√°tica.

‚úÖ La funci√≥n actualizar(frame) cambia los valores de Y, desplazando la onda seno en tiempo real.

‚úÖ FuncAnimation genera 100 frames, actualizando cada 50 milisegundos (interval=50).

‚úÖ Muestra un gr√°fico animado, moviendo la onda seno de manera continua.

üöÄ Este c√≥digo funciona en Jupyter Notebook, VS Code y otras plataformas, pero no en Colab, ya que Colab no soporta animaciones en tiempo real de Matplotlib. Si est√°s en Colab, podr√≠amos usar HTML para mostrar la animaci√≥n.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Crear la figura y el eje
fig, ax = plt.subplots()
x = np.linspace(0, 10, 100)  # Valores del eje X
line, = ax.plot(x, np.sin(x))  # L√≠nea inicial

# Funci√≥n de actualizaci√≥n para la animaci√≥n
def actualizar(frame):
    line.set_ydata(np.sin(x + frame * 0.1))  # Desplaza la onda seno en el tiempo
    return line,

# Configurar la animaci√≥n con FuncAnimation
ani = animation.FuncAnimation(fig, actualizar, frames=100, interval=50)

plt.title("Animaci√≥n de una onda seno")
plt.xlabel("X")
plt.ylabel("Amplitud")
plt.grid(True)
plt.show()
```
