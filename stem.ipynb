```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de una señal con valores fuera del rango del ADC
senal_analogica = np.array([-50, 200, 500, 800, 1100, 1500])  # Algunos valores están fuera del rango

# Definimos el rango del ADC
adc_max = 1023  # Resolución de 10 bits

# Aplicamos np.clip() para restringir los valores al rango permitido del ADC
senal_adc = np.clip(senal_analogica, 0, adc_max).astype(int)

# Gráficos con plt.stem()
plt.figure(figsize=(8, 5))

# Señal original (sin ajuste)
plt.stem(senal_analogica, label='Señal original (con valores fuera del ADC)', linefmt='b-', markerfmt='bo', basefmt=" ")

# Señal ajustada al ADC
plt.stem(senal_adc, label='Señal ajustada para el ADC', linefmt='r-', markerfmt='ro', basefmt=" ")

plt.xlabel("Muestras")
plt.ylabel("Valor")
plt.title("Ajuste de Señal para el ADC")
plt.legend()
plt.grid(True)
plt.show()

# Imprimimos los valores antes y después
print("Valores originales:", senal_analogica)
print("Valores ajustados para el ADC:", senal_adc)
```
