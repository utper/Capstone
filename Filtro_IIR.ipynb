✅ Diseñamos un filtro IIR tipo Butterworth de orden 2 con una frecuencia de corte de 50 Hz.

✅ Generamos una señal senoidal de 10 Hz y le agregamos ruido aleatorio.

✅ Aplicamos el filtro usando filtfilt(), que corrige el desfase del filtro IIR.

✅ Visualizamos la señal ruidosa y la señal filtrada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# 1️⃣ Definir parámetros
fs = 500  # Frecuencia de muestreo (Hz)
fc = 50   # Frecuencia de corte (Hz)
orden = 2 # Orden del filtro IIR

# 2️⃣ Calcular coeficientes del filtro Butterworth IIR
b, a = butter(orden, fc / (fs / 2), btype='low', analog=False)

# 3️⃣ Generar señal ruidosa
t = np.linspace(0, 1, fs, endpoint=False)  # Vector de tiempo
senal_pura = np.sin(2 * np.pi * 10 * t)  # Señal senoidal de 10 Hz
ruido = np.random.normal(0, 0.3, size=t.shape)  # Ruido aleatorio
senal_ruidosa = senal_pura + ruido  # Señal con ruido agregado

# 4️⃣ Aplicar el filtro IIR
senal_filtrada = filtfilt(b, a, senal_ruidosa)

# 5️⃣ Graficar resultados
plt.figure(figsize=(10, 4))
plt.plot(t, senal_ruidosa, label="Señal ruidosa", alpha=0.6, color="gray")
plt.plot(t, senal_filtrada, label="Señal filtrada (IIR)", linewidth=2, color="red")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title(f"Filtro IIR pasa-bajas (fc={fc} Hz)")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
