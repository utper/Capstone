```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
fs = 1000  # Frecuencia de muestreo en Hz
t = np.linspace(0, 1, fs, endpoint=False)  # 1 segundo

# Seal senoidal principal
amplitud_seno = 3.3
frecuencia_seno = 1  # Hz
senoidal = amplitud_seno * np.sin(2 * np.pi * frecuencia_seno * t)

# Seal de ruido (senoidal de alta frecuencia)
amplitud_ruido = 0.3
frecuencia_ruido = 200  # Hz
ruido = amplitud_ruido * np.sin(2 * np.pi * frecuencia_ruido * t)

# Seal resultante
senal_ruidosa = senoidal + ruido

# Graficar
plt.figure(figsize=(12, 6))
plt.plot(t, senoidal, label='Seal Senoidal 1Hz - 3.3V', color = 'blue', alpha=0.7)
plt.plot(t, ruido, label='Ruido 50Hz - 0.3V', color = 'orange', alpha=0.5)
plt.plot(t, senal_ruidosa, label='Seal Ruidosa (Resultado)', color = 'red', linewidth=2)
plt.title("Seal Senoidal con Ruido")
plt.xlabel("Tiempo [s]")
plt.ylabel("Voltaje [V]")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
