```python
# En este código, estás entrenando una red neuronal simple con una capa densa (Dense) que tiene un solo unit y una input_shape de [1]
import sys

import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
# This script requires TensorFlow 2 and Python 3.
if tf.__version__.split('.')[0] != '2':
    raise Exception((f"The script is developed and tested for tensorflow 2. "
                     f"Current version: {tf.__version__}"))

if sys.version_info.major < 3:
    raise Exception((f"The script is developed and tested for Python 3. "
                     f"Current version: {sys.version_info.major}"))
```

#### First lets re-train our original single layer network and see what the prediction is for X = 10.0 and what the learned weights are.

```python
my_layer = keras.layers.Dense(units=1, input_shape=[1])
model = tf.keras.Sequential([my_layer])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
```

```python
print(model.predict(np.array([10.0])))
```

```python
# imprimirá los pesos (weights) y el sesgo (bias) aprendidos por la capa my_layer
# Dado que los datos (xs y ys) siguen una relación lineal de la forma y = 2x - 1, el modelo debería aprender pesos cercanos a 2.0 y un
# sesgo (bias) cercano a -1.0
print(my_layer.get_weights())
```

#### Next lets train a 2-layer network and see what its prediction and weights are.

```python
# Este código crea y entrena una red neuronal en TensorFlow con Keras, utilizando dos capas densas (Dense).
# Crea una capa densa con 2 neuronas y un input de tamaño 1.
my_layer_1 = keras.layers.Dense(units=2, input_shape=[1])
# Crea una segunda capa con 1 neurona, que servirá como salida del modelo.
my_layer_2 = keras.layers.Dense(units=1)

# Creación del modelo y compilación:
# Se define un modelo secuencial que conecta las capas creadas.
model = tf.keras.Sequential([my_layer_1, my_layer_2])
# Se configura el modelo con un optimizador SGD (Gradiente Descendente Estocástico) y una función de pérdida de error cuadrático medio.
model.compile(optimizer='sgd', loss='mean_squared_error')

# Definición de datos de entrenamiento:
# Se crean los datos de entrada y salida, lo que sugiere una relación lineal entre ellos.
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Se entrena el modelo con los datos dados durante 500 épocas.
model.fit(xs, ys, epochs=500)
```

```python
# Se prueba el modelo haciendo una predicción para el valor de entrada 10.0.
print(model.predict(np.array([10.0])))
```

```python
# Se extraen los pesos (w) y los sesgos (b) de my_layer_1 y my_layer_2, lo que permite calcular manualmente la salida.
print(my_layer_1.get_weights())
print(my_layer_2.get_weights())
```

#### Finally we can manually compute the output for our 2-layer network to better understand how it works.

```python
value_to_predict = 10.0

# Extraemos los pesos y sesgos de las capas de red neuronal
layer1_w1 = (my_layer_1.get_weights()[0][0][0])
layer1_w2 = (my_layer_1.get_weights()[0][0][1])
layer1_b1 = (my_layer_1.get_weights()[1][0])
layer1_b2 = (my_layer_1.get_weights()[1][1])

layer2_w1 = (my_layer_2.get_weights()[0][0])
layer2_w2 = (my_layer_2.get_weights()[0][1])
layer2_b = (my_layer_2.get_weights()[1][0])

# Cálculo manual de la salida de las neuronas:
# Se simula el paso de datos a través de la red neuronal de manera manual.
neuron1_output = (layer1_w1 * value_to_predict) + layer1_b1
neuron2_output = (layer1_w2 * value_to_predict) + layer1_b2

neuron3_output = (layer2_w1 * neuron1_output) + (layer2_w2 * neuron2_output) + layer2_b

print(neuron3_output)
```
